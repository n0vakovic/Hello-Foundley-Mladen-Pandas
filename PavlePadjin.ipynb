```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(1,100,100)
y1 = pd.Series(5*x+7)
y2 = pd.Series(5*x+7 +  25 *np.sin(x))
df = {"x":pd.Series(x), "expected":y1, "reality":y2}
df = pd.DataFrame(df)
plt.plot(df['x'],df['expected'])
plt.plot(df['x'],df['reality'])
plt.legend(['Expected progress rate','Probable progress rate'])
plt.title('Expected VS Most likely progress rate while working with Mladen')
plt.xlabel('Time')
plt.ylabel('Progress rate')
```

I gotta admit, I heard about Foundley from a friend at first and I didn't expect much. But **now** I see that it represents an ocean of posibilities. 

```python
df['x-small'] = df['x'][:10]
df['foundley'] = np.exp(df['x-small'])
plt.plot(df['x-small'],df['foundley'])
plt.title('My appreciation for Foundley over time')
plt.xlabel('Time')
plt.ylabel('Appreciation for Foundley')
```
